## CREDIT CARD FRAUD DETECTION

```python
#importing the required libraries and packages for the project.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Pandas is a python library used for working with dataframes.<br>
Matplotlib is a data visualization and a graph plotting library.<br>

```python
df=pd.read_csv("creditcard.csv")
df.head()
```

```python
#
df.info()
```

```python
df['Class'].value_counts()
```

The given dataset is highly unbalanced.<br>
0--> Normal Transaction<br>
1--> Fraudulent Transaction

```python
normal = df[df.Class == 0]
fraud = df[df.Class == 1]
```

```python
normal.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
df.groupby('Class').mean()
```

UNDERFITTING<br>
We will be building a sample dataset containing similar distribution of normal and fraudulent transactions<br>
There are 492 fraudulent transactions.

```python
normal_sample = normal.sample(n=492)
```

```python
new_df = pd.concat([normal_sample, fraud], axis=0)
new_df
```

```python
new_df.Class.value_counts()
```

```python
new_df.groupby('Class').mean()
```

SPLITTING DATA INTO FEATURES AND TARGETS

```python
X = new_df.drop(columns='Class', axis=1)
Y = new_df['Class']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=13)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

TRAINING MODEL

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

MODEL EVALUATION

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
training_data_accuracy
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
test_data_accuracy
```
