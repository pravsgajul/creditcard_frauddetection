## CREDIT CARD FRAUD DETECTION

```python
#importing the required libraries and packages for the project.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Numpy is a python library that is used for working with arrays.<br>
Pandas is a python library used for working with dataframes.<br>
Matplotlib is a data visualization and a graph plotting library.<br>
Seaborn is a library used for data visualization and exploratory data analysis<br>
Scikitlearn is a library algorithm that features multiple Machine Learning algorithms.

```python
# Importing the required dataset
df=pd.read_csv("creditcard.csv")
df.head()
```

```python
#Checking the count and the data type and the number of non null values
df.info()
```

```python
#Finding the distribution of the given data
df['Class'].value_counts()
```

The given dataset is highly unbalanced.<br>
0 refers to Normal Transaction<br>
1 refers to Fraudulent Transaction

```python
#Converting the numerical data into two different dataframes by the names of normal and fraud
normal = df[df.Class == 0]
fraud = df[df.Class == 1]
```

```python
#Describing normal dataframe's amount attribute
normal.Amount.describe()
```

```python
#Describing fraud dataframe's amount attribute
fraud.Amount.describe()
```

```python
#Grouping the data by class attribute
df.groupby('Class').mean()
```

There exists underfitting in the data due to 492 fraudulent transactions<br>
We will be building a sample dataset containing similar distribution of normal and fraudulent transactions<br>

```python
#Creating sample dataset
normal_sample = normal.sample(n=492)
```

```python
#combining normal_sample and fraud dataframes
new_df = pd.concat([normal_sample, fraud], axis=0)
new_df
```

```python
#Checking balances of classes
new_df.Class.value_counts()
```

```python
#grouping by class attribute
new_df.groupby('Class').mean()
```

SPLITTING DATA INTO FEATURES AND TARGETS

```python
#Splitting data into testing and training sets
X = new_df.drop(columns='Class', axis=1)
Y = new_df['Class']
```

```python
#Using test train split from scikit learn
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=13)
```

```python
#Shapes of the split data
print(X.shape, X_train.shape, X_test.shape)
```

TRAINING MODEL

```python
#Building a logistic regression model
model = LogisticRegression()
```

```python
# training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

MODEL EVALUATION

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
training_data_accuracy
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
test_data_accuracy
```
